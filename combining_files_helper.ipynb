```python
import pandas as pd
```

```python
files = {
    "CA": "data/california_insights.csv",
    "TX": "data/texas_insights.csv",
    "FL": "data/florida_insights.csv",
    "WA": "data/washington_insights.csv",
    "NY": "data/new_york_insights.csv",
    "VA": "data/virginia_insights.csv",
    "VT": "data/vermont_insights.csv"
}

dfs = []

for state, file in files.items():
    df = pd.read_csv(file)
    df["state"] = state
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)

combined_df.to_csv("combined_insights.csv", index=False)

print("Combined dataset created.")
```
