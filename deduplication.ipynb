```python
import os
import pandas as pd

def process_all_datasets(folder_path):
    """
    Process all CSV files in a folder, remove duplicates by gmap_id, and save cleaned versions.
    
    Parameters:
    -----------
    folder_path : str
        Path to folder containing CSV files
    """
    # Get all CSV files in folder
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    
    if not csv_files:
        print(f"No CSV files found in {folder_path}")
        return
    
    for file in csv_files:
        file_path = os.path.join(folder_path, file)
        
        try:
            # Read CSV
            df = pd.read_csv(file_path)
            original_count = len(df)
            
            # Remove duplicates by gmap_id
            df_cleaned = df.drop_duplicates(subset=['gmap_id'], keep='first')
            cleaned_count = len(df_cleaned)
            
            # Overwrite original file
            df_cleaned.to_csv(file_path, index=False)
            
            duplicates_removed = original_count - cleaned_count
            print(f"✓ {file}: {original_count} → {cleaned_count} rows ({duplicates_removed} duplicates removed)")
            
        except Exception as e:
            print(f"✗ {file}: Error - {e}")

# Usage:
process_all_datasets('Data/Final_Cus_Bus_data')
```
