```python
import gzip
import json
import pandas as pd

def parse(path):
  g = gzip.open('data/' + path, 'r')
  for l in g:
    yield json.loads(l)

def parse_reviews(path, N = 1_000_000):
  g = gzip.open('data/' + path, 'r')
  for i, l in enumerate(g):
    if i >= N:
      break
    yield json.loads(l)
```

```python
cali_meta = parse('meta-California.json.gz')
cali_meta_table = pd.DataFrame(cali_meta)
```

```python
cali_reviews = parse_reviews('review-California.json.gz', )
cali_reviews_table = pd.DataFrame(cali_reviews)
```
